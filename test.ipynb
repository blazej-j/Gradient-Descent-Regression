```python
from generator import DataGenerator
from visualize import Visualizer
from model import RegressionModel

import math_functions as mf
import loss_functions as lf
```

```python
dg = DataGenerator(mf.linear_fn)
x, y = dg.generate(npoints=150)

rm = RegressionModel(mf.linear_fn,lf.mae)
rm.fit(x,y, epochs=100)

a = Visualizer(generator=dg, model=rm)
a.plot_training_overview()
a.plot_error()

rm.score(x,y)
```
